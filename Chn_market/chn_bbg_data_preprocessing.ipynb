```python
def lemmatize_words(sentance):

    sentance = sentance.lower()
    tokenizer = RegexpTokenizer(r"\w+")
    words = tokenizer.tokenize(sentance)
    lemmatized_words = [
        WordNetLemmatizer().lemmatize(word, "v") for word in words if word != ""
    ]

    return lemmatized_words


intro_pd["lemma"] = intro_pd["intro"].apply(lemmatize_words)
intro_pd["intro_lemmatized"] = intro_pd["lemma"].map(" ".join)
intro_pd['company']=intro_pd['company'].map(lambda x: x+'.SH' if (x[0]=='6' or x[0]=='9') else x+'.SZ')
intro_pd.head()
```

```python
desc = pd.read_csv("data/bbg_company_des_chinese_ch.csv",names=['company','intro'],dtype=str)
def is_contains_chinese(strs):
    for _char in strs:
        if '\u4e00' <= _char <= '\u9fa5':
            return True
    return False
desc=desc[desc['intro'].map(is_contains_chinese)]
desc['company']=desc['company'].map(lambda x: x+'.SH' if (x[0]=='6' or x[0]=='9') else x+'.SZ')
desc['length']=desc.intro.map(len)
desc
```
